In this tutorial, we are going to take a look at one of the most fundamental regression techniques called the <b>Linear Regression</b> with only one parameter.

### Libraries & Files

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
housing = pd.read_csv('housing.csv')
housing = housing[['price', 'area']]
```

### What is Linear Regression?

The fundamental idea of Machine Learning is to let machines "learn" to perform certain tasks.

Imagine the following scenario: you are helping your friend John to sell his house. However, you are not quite sure what the house is worth. In this case, you might want to take some <b>features</b> of the house into consideration to <b>predict</b> the house price. To make things simple, let's say that you are only considering the <b>area</b> of the house.

Essentially, we are using the <b>house size</b> to predict the <b>house price</b>. Let's denote these variables as following:

|Notation  <img width=70/> | Description<img width=350/>| Python |
|: ------------|: ------------------------------------------------------------||
| $x_{i}$ | The ith house size in our dataset | `x_i`
| $y_{i}$ | The ith house price prediction based on $x_{i}$d | `y_i`

### What Does This Look Like?

Now, let's say we plot this data and get the following plot:

```python
plt.scatter(housing['area'], housing['price'])

plt.title("Housing Prices")
plt.xlabel("Area (in square feet)")
plt.ylabel("Price (in USD millions)")
```

From this plot above, we can see that there is a <b>positive linear relationship</b> across the points, meaning we can draw a linear line in the form of <b>y = ax + b</b>.

### How Can We Train Our Model From This?

With this information in mind, we will use the 0 to 400th data points to <b>train</b> our model, which basically means to <b>provide materials so that the machine learning model can learn a pattern</b>.

```python
x_train = np.array(housing['area'][0:450]).reshape((-1, 1)) # x has to be two-dimensional (i.e. one column and multiple rows)
y_train = np.array(housing['price'][0:450])

x_test = np.array(housing['area'][450:]).reshape((-1, 1))
y_test = np.array(housing['price'][450:])
```

```python
lg_model = LinearRegression()
```

```python
lg_model.fit(x_train, y_train)
```

Now, we have this model called `lg_model` that learned all of the 400 data points that we provided. Now, we can call the following line to find the slope `a` and intercept `b` from <b>y = ax + b</b> we defined earlier.

```python
slope = lg_model.coef_
intercept = lg_model.intercept_
print((slope[0], intercept))
```

Then, we can plot this line to show whether it looks right.

```python
plt.scatter(housing['area'], housing['price'])
plt.title("Housing Prices")
plt.xlabel("Area (in square feet)")
plt.ylabel("Price (in USD millions)")

prediction = slope[0]*housing['area'].values + intercept
plt.plot(housing['area'], prediction, color='red')
```

### Performance?

Finally, we can find the <b>performance</b> on the rest of the data by running the following lines of code.

```python
train_score = lg_model.score(x_train, y_train)
train_score
```

```python
test_score = lg_model.score(x_test, y_test)
test_score
```
